```python
# Define styles and visual properties for each dataset
style_map = {
    "ani1": {"label": "Space (ANI)", "color": "C0", "marker": "s", "style": "-"},
    "ani4": {"label": "ANI-1", "color": "C0", "marker": "s", "style": "--"},
    "qm1": {"label": "Space (QM9)", "color": "C1", "marker": "^", "style": "-"},
    "qm4": {"label": "QM9", "color": "C1", "marker": "^", "style": "--"},
    "gdb1": {"label": "Space (GDB)", "color": "C2", "marker": "o", "style": "-"},
    "gdb4": {"label": "GDB-13", "color": "C2", "marker": "o", "style": "--"},
}

# Provide total space sizes for normalisation of sample fraction (x-axis)
space_sizes = {
    "ani1": 3.663e+06, # Total space size for ANI-1
    "qm1": 5.897e+07,
    "gdb1": 1.100e+15,

    "ani4": 57462, # Total size for ANI-1 database
    "qm4": 130831,
    "gdb4": 977468254,
}

# Dictionary containing all relevant DataFrames
all_dfs = {
    "ani1": ani1,
    "ani4": ani4,
    "qm1": qm1,
    "qm4": qm4,
    "gdb1": gdb1,
    "gdb4": gdb4,
}
```

```python
#Create figure and axis
fig, ax = plt.subplots(figsize=(10, 3))

#Remove top and right spines for a cleaner look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#Iterate through each dataset and plot its KS values with error bars
for key, props in style_map.items():
    df = all_dfs[key]

    #Compute the sample fraction by dividing sample count by total space size
    x = df.index.astype(float) / space_sizes[key]
    y = df["KS_Mean"].astype(float)

    # Determine the error bars based on available columns
    if "KS_CI_95" in df.columns:
        ci = df["KS_CI_95"].str.strip("[]").str.split(",", expand=True).astype(float)
        yerr = [
            (y - ci[0]).clip(lower=0),
            (ci[1] - y).clip(lower=0)
        ]
    elif "KS_Low" in df.columns and "KS_High" in df.columns:
        yerr = [
            (y - df["KS_Low"].astype(float)).clip(lower=0),
            (df["KS_High"].astype(float) - y).clip(lower=0)
        ]
    else:
        yerr = None

    # Plot the data with error bars
    ax.errorbar(
        x, y,
        fmt=props["marker"],
        label=props["label"],
        color=props["color"],
        linestyle=props["style"],
        capsize=3,
        yerr=yerr
    )

#Plot settings
ax.set_xscale("log")
ax.set_xlabel("Sample fraction", fontsize=12)
ax.set_ylabel("KS", fontsize=12)
ax.legend(fontsize="small", ncol=2)
plt.tight_layout()

ax.legend(
    fontsize="small",
    ncol=2,
    loc="upper left",
    bbox_to_anchor=(0.07, 1.0)  )



plt.savefig("ks_plot.pdf", format="pdf", bbox_inches="tight") #to save the plot as a PDF
plt.show()
```
